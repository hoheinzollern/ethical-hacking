# Stack-based VMs

- e.g. JVM/Java, Python etc.
- bytecode maps more directly to source language
- easier decompilation: [Ghidra](https://ghidra-sre.org/), [Mocha](http://www.brouhaha.com/~eric/software/mocha/)

# Register-based VMs

- e.g. .NET/CLR, Dalvik/Java etc.
- infinite number of registers available, unlike real CPUs
- also easy to decompile: [dotPeek](https://www.jetbrains.com/decompiler/), [JEB](https://www.pnfsoftware.com/)
